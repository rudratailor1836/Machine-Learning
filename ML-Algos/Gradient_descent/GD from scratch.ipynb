```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
X,y = make_regression(n_samples=100,n_features=1, n_informative=1, n_targets=1, noise=20)
```

```python
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X,y)
print(lr.coef_)
print(lr.intercept_)
```

```python
class GDRegressor:
    
    def __init__(self,slope,learning_rate, epochs):
        self.m = slope
        self.b = -120
        self.lr = learning_rate
        self.epochs = epochs 
        
    def fit(self,X,y):
        for i in range(self.epochs):
            loss_slope = -2*np.sum(y - self.m*X.ravel() - self.b)
            self.b = self.b - (self.lr * loss_slope)
        print(self.m)
        print(self.b)
    
    
            
            
```

```python
gd = GDRegressor(88.59,0.001,100)
```

```python
gd.fit(X,y)
```
