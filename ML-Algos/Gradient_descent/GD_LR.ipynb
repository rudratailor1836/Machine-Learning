```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=80, random_state=13)
```

```python
import matplotlib.pyplot as plt 
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')
```

```python
# Lets apply GD assuming slope is constant m = 78.35
# and lets assume the starting val for intercept is b = 0
```

```python
y_pred = ((78.35* X) + 0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red', label='OLs')
plt.plot(X,y_pred,color='black',label = 'b = 0')
plt.legend()
plt.show()
```

```python
m = 78.35
b = -1000

loss_slope = -2 * np.sum(y- m*X.ravel() - b)
loss_slope
```

```python
lr = 0.1

step_size = loss_slope * lr
step_size
 
```

```python
b =  b - step_size
b
```

```python
y_pred_new = ((78.35 * X) + b).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red', label='OLs')
plt.plot(X,y_pred_new,color='green',label = f'b = {b}')
plt.plot(X,y_pred,color='black',label = 'b = 0')
plt.legend()
plt.show()
```
