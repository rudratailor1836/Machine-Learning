```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 

import scipy.stats as stats 
```

```python
from sklearn.model_selection import train_test_split as tts
from sklearn.model_selection import cross_val_score 

from sklearn.linear_model import LinearRegression 

from sklearn.metrics import r2_score

from sklearn.preprocessing import PowerTransformer 
```

```python
df = pd.read_csv('concrete_data.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
X = df.drop(columns=['Strength'])
y = df.iloc[:,-1]
```

```python

X_train, X_test, y_train, y_test = tts(X,y,test_size=0.2,random_state=42)
```

```python
# Without Transformation
lr = LinearRegression()

lr.fit(X_train,y_train)

y_pred = lr.predict(X_test)

r2_score(y_test,y_pred)
```

```python
lr = LinearRegression()
np.mean(cross_val_score(lr,X,y,scoring='r2'))
```

```python
# plotting the distplot without any transformation

for col in X_train.columns:
    plt.figure(figsize=(14,4))
    sns.distplot(X_train[col])
    plt.title(col)
    
    plt.subplot(122)
    stats.probplot(X_train[col], dist='norm', plot = plt)
    plt.title(col)
    
    plt.show()
```

```python
#  Applying Box-Cox

pt = PowerTransformer(method='box-cox')

X_train_trn = pt.fit_transform(X_train+0.000001)
X_test_trn = pt.transform(X_test+0.000001)

pd.DataFrame({'cols':X_train.columns,'box_cox_lambdas': pt.lambdas_})
```

```python
lr = LinearRegression()
lr.fit(X_train_trn,y_train)

y_pred1 = lr.predict(X_test_trn)

r2_score(y_test, y_pred1)
```

```python
# using cross val score
pt  = PowerTransformer(method='box-cox')
X_trns = pt.fit_transform(X+0.000001)

lr= LinearRegression()
np.mean(cross_val_score(lr,X_trns,y,scoring='r2'))
```

```python
X_train_trn = pd.DataFrame(X_train_trn,columns=X_train.columns)
for col in X_train_trn.columns:
    plt.figure(figsize=(14,4))
    plt.subplot(121)
    sns.distplot(X_train[col])
    plt.title(col)
    
    plt.subplot(122)
    sns.distplot(X_train_trn[col])
    plt.title(col)
    
    plt.show()
```

```python
# Yeo Johnson transform
pt1 = PowerTransformer()

X_train_trn1 = pt1.fit_transform(X_train)
X_test_trn1 = pt1.transform(X_test)

lr = LinearRegression()
lr.fit(X_train_trn1,y_train)

y_pred2 = lr.predict(X_test_trn1)

r2_score(y_test,y_pred2)
```
