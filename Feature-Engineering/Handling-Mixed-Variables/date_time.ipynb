```python
# converting to datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

Working with Date

```python
date.info()
```

```python
# Extracting year, month, day, weekday
date['year'] = date['date'].dt.year
date['month'] = date['date'].dt.month
date['month_name'] = date['date'].dt.month_name()
date['day'] = date['date'].dt.day
date['weekday'] = date['date'].dt.dayofweek
date['weekday_name'] = date['date'].dt.day_name()


date.sample(7)
```

```python
# is weekend? 
date['weekend'] = np.where(date['weekday_name'].isin(['Sunday','Saturday']), True,False)

date.sample(7)
```

```python
# Extracting week of the y
date = date.drop(columns=['date_week'])

date.sample(7)
```

```python
# Extracting quatrer
date['quarter']= date['date'].dt.quarter


date.sample(7)
```

```python
# extracting Semester
date['semester'] = np.where(date['quarter'].isin([1,2]), 1,2)


date.sample(7)
```

```python
# Extract Time elapsed between dates

import datetime 

today = datetime.datetime.today()

today
```

```python
(today - date['date']).dt.days
```

```python
# Months Passed
np.round((today - date['date']) / np.timedelta64(1, 'D'),0)
```
