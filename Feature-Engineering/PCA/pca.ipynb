```python
import numpy as np 
import pandas as pd
```

```python
np.random.seed(23)

mu_vec1 = np.array([0,0,0])
cov_mat1 = np.array([[1,0,0],[0,1,0],[0,0,1]])
class1_sample = np.random.multivariate_normal(mu_vec1, cov_mat1, 20)

df = pd.DataFrame(class1_sample, columns=['feature1','feature2','feature3'])
df['target'] = 1

mu_vec2 = np.array([1,1,1])
cov_mat2 = np.array([[1,0,0],[0,1,0],[0,0,1]])
class2_sample = np.random.multivariate_normal(mu_vec2, cov_mat2, 20)

df1 = pd.DataFrame(class2_sample, columns=['feature1','feature2','feature3'])
df1['target'] = 0

combined_df = pd.concat([df,df1],ignore_index=True)

combined_df = combined_df.sample(40)
combined_df.head()
```

```python
mdf = combined_df
```

```python
import plotly.express as px
```

```python


# y_train_trf = y_train.astype(str)
fig = px.scatter_3d(mdf, x='feature1', y='feature2', z='feature3',
                    color=mdf['target'].astype('str'))
fig.update_traces(marker=dict(size=12,
                              line=dict(width=2,
                                        color='grey')),
                  selector=dict(mode='markers'))
```

```python
# step1 mean standarding

from sklearn.preprocessing import StandardScaler 
s1 = StandardScaler()

df.iloc[:,0:3] = s1.fit_transform(df.iloc[:,0:3])

# step2 finding Covariance matrix
covariance_matric = np.cov([df.iloc[:,0],df.iloc[:,1],df.iloc[:,2]])
print(covariance_matric)
```

```python
eigen_values, eigen_vectors = np.linalg.eig(covariance_matric)

eigen_values
```
